```python
import matplotlib.pyplot as plt

student = ['ayush', 'piyush', 'ved', 'hemant', 'ravi']
percentage = [95, 85, 25, 45, 75]

plt.title('student result of 2025')

plt.bar(student, percentage, color='green', label='student grade')

plt.xlabel('Student Name')
plt.ylabel('Marks')

plt.xticks(['ayush', 'piyush', 'ved', 'hemant', 'ravi'], ['Ayush', 'Piyush', 'Ved', 'Hemant', 'Ravi'])
plt.yticks([0, 20, 40, 60, 80, 100])
plt.grid()
plt.legend()

plt.show()
```
