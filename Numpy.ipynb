#   Differ between List and Numpy

**. Time Taken through List**

```python
import time
Start = time.time()

list1 = [i for i in range(10000000)]
list2 = [i for i in range(10000000, 20000000)]
list3 = []
for i in range(len(list1)):
    list3.append(list1[i] + list2[i])
print(f"Tike taken Through list = ", time.time()-Start)
```

***. Time Taken through Numpy***


```python
import numpy as np 
import time

Start = time.time()

list1 = np.arange(10000000)
list2 = np.arange(10000000,20000000)

list3 = list1+list2

print(f"Tike taken Through Numpy = ", time.time()-Start)
```
