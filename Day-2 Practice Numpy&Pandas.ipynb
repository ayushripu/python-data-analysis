# Pandas

```python
import pandas as pd
```

```python
a = pd.DataFrame(
    {
        'Name' : ['a','b','c','a','b'],
        'Age' : [20, 22, 21, 20, 22]
    }
)
a
```

```python
a.head(2)
```

```python
a[['Age']]
```

```python
a[a['Age']>21]
```

```python
a['Status'] = 'junior'
a.loc[a['Age']>21, 'Status'] = 'Senior'
a
```

```python
a.sort_values(by = 'Age', ascending= False)
```

```python
a['Age'].mean()
```

```python
a['Age'].value_counts()
```

```python
a.groupby('Age').count()
```

```python
df = pd.DataFrame({
    'Dept': ['IT','HR','IT','Finance','HR'],
    'Salary': [50000,40000,60000,45000,42000]
})
df
```

```python
df.groupby('Dept')['Salary'].sum()
```

```python
df.groupby('Dept')['Salary'].mean()
```
