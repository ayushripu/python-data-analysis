#                                 Matplotlib

```python
import matplotlib.pyplot as plt

x=  [1,2,3,4]
y = [10,20,-10,25]

plt.plot(x,y)
plt.show()
```

# Matplotlib ploting function

**1. plot()**

```python
import matplotlib.pyplot as plt

month = [1,2,3,4]
sales = [456, 128, 500, 120]

plt.plot(month, sales)
plt.show()
```

```python
import matplotlib.pyplot as plt

month = [1,2,3,4]
sales = [456, 128, 500, 120]

plt.plot(month,sales,  color = 'red')
plt.show()

```

```python
import matplotlib.pyplot as plt

month = [1,2,3,4]
sales = [456, 128, 500, 120]

plt.plot(month,sales,  color = 'red',marker = '.')
plt.show()



```

```python
import matplotlib.pyplot as plt

month = [1,2,3,4]
sales = [456, 128, 500, 120]

plt.plot(month,sales,  color = 'red',marker = '.', linestyle = ':')
plt.show()



```

**2. Label**

```python
import matplotlib.pyplot as plt

month = [1,2,3,4]

sales = [456,128,500,120]
profit = [100,50,200,80]

plt.plot(month, sales, linestyle = ':', label='sales',  color = 'g')
plt.plot(month, profit, linestyle = '--', label='profit', color = 'r')



plt.legend()
plt.show()



```

**3. Label and Legend()**

```python
import matplotlib.pyplot as plt

month = [1,2,3,4]

sales = [456,128,500,120]
profit = [100,50,200,80]

plt.plot(month, sales, linestyle = ':', label='sales',  color = 'g')
plt.plot(month, profit, linestyle = '--', label='profit', color = 'r')



plt.legend(loc = 'upper right')
plt.show()



```

**4. Title()**

```python
import matplotlib.pyplot as plt

month = [1,2,3,4]

sales = [456,128,500,120]
profit = [100,50,200,80]

plt.title("Monthly Sales vs Profit")

plt.plot(month, sales, linestyle = ':', label='sales',  color = 'g')
plt.plot(month, profit, linestyle = '--', label='profit', color = 'r')

plt.xlabel('Month')
plt.ylabel('Profit and sales')

plt.legend()
plt.show()



```

**5. Grid()**

```python
import matplotlib.pyplot as plt

month = [1,2,3,4]

sales = [456,128,500,120]
profit = [100,50,200,80]

plt.plot(month, sales, marker = '.', linestyle = '-', label='sales',  color = 'g')
plt.plot(month, profit, marker = '*', linestyle = '--', label='profit', color = 'r')

plt.legend()

plt.grid(color = 'yellow', linestyle= ':', linewidth = 1)

plt.show()



```

**6.Axis Limits**

```python
import matplotlib.pyplot as plt

month = [1,2,3,4]

sales = [456,128,500,120]
profit = [100,50,200,80]

plt.plot(month, sales, marker = '.', linestyle = '-', label='sales',  color = 'g')
plt.plot(month, profit, marker = '*', linestyle = '--', label='profit', color = 'r')

plt.legend()

plt.grid(color = 'yellow', linestyle= ':', linewidth = 1)

plt.xlim(1,3)
plt.ylim(100,300)
plt.show()



```

**7.  Axis Ticks**

**plt.xticks()**  -- change the defalut value of X-axis like [1,2,3,4,5] 

**plt.yticks()**  -- change the defalut value of Y-axis like [100,200,300,400,500] 

```python
import matplotlib.pyplot as plt

month = [1,2,3,4]

sales = [456,128,500,120]
profit = [100,50,200,80]

plt.plot(month, sales, marker = '.', linestyle = '-', label='sales',  color = 'g')
plt.plot(month, profit, marker = '*', linestyle = '--', label='profit', color = 'r')

plt.legend()

plt.grid(color = 'yellow', linestyle= ':', linewidth = 1)

plt.xticks([1,2,3,4], ['Month1', 'Month2', 'Month3', 'month4'])

plt.show()



```
