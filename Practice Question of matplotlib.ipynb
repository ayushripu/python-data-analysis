# Q1 Plot a simple line chart for:

X = [1, 2, 3, 4, 5]
Y = [10, 20, 15, 25, 30]

👉 Add:

Title

X & Y labels


```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5] 
y = [10, 20, 15, 25, 30]

plt.plot(x,y)
plt.title('Plot line Chart')
plt.xlabel('Day in week')
plt.ylabel('Sales in week')

plt.show()
```

# Q2. Create a line plot with custom styling:

Change color to red

Use dashed line (--)

Add markers (o)

```python
month = [1,2,3,4,5]
sales = [50000,45000,12000,10000,5000]

plt.plot(month, sales, color = 'red', marker = 'o', linestyle = '--')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')


plt.show()
```

# Q3. Plot two lines in same graph:

Sales = [100, 200, 300, 400]

Profit = [20, 50, 80, 120]

👉 Add legend

```python
Sales = [100, 200, 300, 400]

Profit = [20, 50, 80, 120]

Day = [1, 2, 3, 4] 

plt.plot(Day, Sales, label = 'sales', marker = 'o')
plt.plot(Day, Profit, label = 'Profit', marker = '*')

plt.title('Sales & Profit')
plt.xlabel('Day')
plt.ylabel('Amount')

plt.legend()
plt.grid(True)

plt.show()
```

# Q4. Create a bar chart for:

Categories: ['A', 'B', 'C', 'D']

Values: [5, 7, 3, 8]

```python
Categories= ['A', 'B', 'C', 'D']

Values= [5, 7, 3, 8]

plt.bar(Categories, Values, color='skyblue')


plt.title('Category-wise Values')
plt.xlabel('Categories')
plt.ylabel('Values')

plt.grid(axis='y')

plt.show()
```

# Q5. Create a horizontal bar chart

```python
Categories= ['A', 'B', 'C', 'D']

Values= [5, 7, 3, 8]

plt.barh(Categories, Values, color='skyblue')


plt.title('Category-wise Values')
plt.xlabel('Values')      
plt.ylabel('Categories')  

plt.grid(axis='x')

plt.show()
```

# Q6. Create a histogram:

Generate 100 random numbers

Show distribution

Add bins = 10

```python
import random

data= [random.randint(1,100) for _ in range(100)]

plt.hist(data, bins=10, edgecolor = 'black')

plt.title('Random Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```

# Q7. Create a pie chart:

Data: [40, 30, 20, 10]

Labels: ['Python', 'Java', 'C++', 'JS']

👉 Add percentage display (autopct)

```python
Data= [40, 30, 20, 10]

plt.pie(Data, labels= ['Python', 'Java', 'C++', 'JS'], autopct = '%1.1f%%')
 
plt.title('Subject Coveed')
plt.show()
```

# 8. Create a scatter plot:

X = [1,2,3,4,5]

Y = [2,4,1,8,7]

👉 Change color & size

```python
x = [1,2,3,4,5]

y = [2,4,1,8,7]

plt.scatter(x,y, color = 'red', s=100, marker = '*')

plt.title('Scatter Plot Example')
plt.xlabel('X values')
plt.ylabel('Y values')

plt.show()
```

```python
x = [1,2,3,4]
y = [10,20,15,25]

categories = ['A','B','C','D']
values = [5,7,3,8]

plt.subplot(1, 2, 1)
plt.plot(x,y, marker = 'o')
plt.title('Line Plot')

plt.subplot(1,2,2)
plt.bar(categories, values, color = 'orange', edgecolor = 'black')
plt.title('Bar Chart')

plt.tight_layout()
plt.show()
```

Grid enabled
