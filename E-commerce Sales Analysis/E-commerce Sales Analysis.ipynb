### Step 7: Data Type Conversion

In this step, we convert columns to their appropriate data types to ensure accurate analysis and avoid inconsistencies.

- Converted 'Date' column to datetime format for time-based analysis.
- Converted 'ship-postal-code' to string as it represents an identifier, not a numeric value.
- Converted 'currency' to category to optimize memory usage and improve performance.

```python
df['Date'] = pd.to_datetime(df['Date'], format='%m-%d-%Y', errors='coerce')
```

```python
df['ship-postal-code'] = df['ship-postal-code'].astype(str)
```

```python
df['currency'] = df['currency'].astype('category')
```
