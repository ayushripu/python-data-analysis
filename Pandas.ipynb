# 2. Basic DataFrame Understanding

**. I want top row of data**

```python
Employee = {
                "Name" : ["Ramesh", "Suresh", "Rahu", "Kaju", "Umesh", "Rohan","Suresh", "Ravi"],
                "Emp_id":[101, 102, 103, 104, 105, 106, 107, 108],
                "Salary":[85000, 69310, 48612, 45962, 75961, 45861, 32569, 65897]
           }
df = pd.DataFrame(Employee)

print(df.head(2))   #bydefault it takes top 5 rows
```

**. I want bottom row of data**

```python
Employee = {
                "Name" : ["Ramesh", "Suresh", "Rahu", "Kaju", "Umesh", "Rohan","Suresh", "Ravi"],
                "Emp_id":[101, 102, 103, 104, 105, 106, 107, 108],
                "Salary":[85000, 69310, 48612, 45962, 75961, 45861, 32569, 65897]
           }
df = pd.DataFrame(Employee)

print(df.tail(3))   #bydefault it takes bottom 5 rows
```
