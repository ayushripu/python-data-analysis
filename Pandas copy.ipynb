# 10. Concatenate and Merge(Join) Data Frame

```python
df1= pd.DataFrame({ 'id' : [5,9,6],
                    'Name' : ['Ayush','Piyush','Ankur'],
                   'age' : [23, 25, 35]})
df1
```

```python
df2= pd.DataFrame({'id' : [5,9,6],
                   'MArks' : [48, 95, 19]})
df2
```

```python
pd.concat([df1, df2],axis=0)  #Verticle / row level / top on top
```

```python
pd.concat([df1, df2],axis=1)   #horizontal / column level / side on side
```

```python
#Join

pd.merge(df1, df2, how='inner', on='id')
```

```python
pd.merge(df1, df2, how='left', on='id')
```

```python
pd.merge(df1, df2, how='right', on='id')
```

```python
pd.merge(df1, df2, how='outer', on='id')
```
