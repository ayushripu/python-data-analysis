**Q1: Create DataFrame**


```
Create a DataFrame with:
Name: A, B, C
Marks: 80, 90, 85
```

````python
import pandas as pd
````

````python
df  = pd.DataFrame({'Name' : ['A','B','C'],
                    'Marks': [80,  90, 85]})

df
````

**Q2. Print first 2 rows of DataFrame**

````python
df.head(2)
````

**Q3. Select only "Marks" column**

````python
df[['Marks']]
````

**Q4.  Show students with marks greater than 85**

````python
df[df['Marks']>85]
````


***Q5. Add new column "Grade" based on marks:***
~~~
Marks > 85 → "A"
else → "B"
~~~

````python
df['Grade'] = 'B'
df.loc[df['Marks']>85, 'Grade']='A'
df
````

````python

df['Grade2'] = 'B'
df.loc[df['Marks'] >= 85, 'Grade2']='A'
df
````
